```python
from tqdm import tqdm

import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_step_infinite(num_sims: int, p_right: float, num_steps: int):
    res = np.zeros(num_sims)
    x = np.random.rand(num_steps, num_sims)

    move_full = (x < p_right).astype(int) - (x >= p_right).astype(int)

    res = np.sum(move_full, axis=0)

    return res
```

```python
def random_step_infinite_path(p_right: float, num_steps: int):
    res = np.zeros(1)
    x = np.random.rand(num_steps, 1)

    move_full = (x < p_right).astype(int) - (x >= p_right).astype(int)
    path = list()

    for row in move_full:
        res += row
        path.append(res[0])

    return res, path
```

```python
random_step_infinite_path(0.5, 500)
```

```python
def random_step_finite(
    num_sims: int, p_right: float, num_steps: int, min_: int = -25, max_: int = 25
):
    res = np.zeros(num_sims)
    x = np.random.rand(num_steps, num_sims)

    move_full = (x < p_right).astype(int) - (x >= p_right).astype(int)

    for move in tqdm(move_full):
        res = res + move

        bounds = (res > max_).astype(int) - (res < min_).astype(int)
        res = res - bounds

    return res
```

```python
res_odd = random_step_finite(100_000_0, 0.5, 499)

plt.hist(res_odd, color="skyblue", edgecolor="black", bins=[i - 25.5 for i in range(0, 51)])

# Adding labels and title
plt.xlabel("Position")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
res_even = random_step_finite(100_000_0, 0.5, 500)

plt.hist(res_even, color="skyblue", edgecolor="black", bins=[i - 25.5 for i in range(0, 51)])

# Adding labels and title
plt.xlabel("Position")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```
