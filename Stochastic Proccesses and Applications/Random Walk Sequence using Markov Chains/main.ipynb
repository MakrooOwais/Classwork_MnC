```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 8
p = 0.5

idxs = np.linspace(0, n+1, n+1)
probabil = [0.5] + [0.5/8] * 8
alpha = np.zeros((n+1, 1))
idx = int(np.random.choice(idxs, p=probabil))
print(idx)
alpha[idx] = 1

one_step = np.zeros((n + 1, n + 1))
# one_step[0, 1] = p
one_step[0, 0] = 1
# one_step[-1, -2] = 1 - p
one_step[-1, -1] = 1
for i in range(1, n):
    one_step[i, i-1: i+2] = np.array([1-p, 0, p])
```

```python
one_step
```

```python
probs = (alpha.T @ np.linalg.matrix_power(one_step, 1)).squeeze()
X = np.linspace(0, n, n+1)
plt.bar(X, probs)
```
