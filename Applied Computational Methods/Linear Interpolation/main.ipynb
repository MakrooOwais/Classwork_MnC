```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def func(X: np.ndarray | float):
    return 2* np.exp(X) - 5 * np.sin(X)
```

```python
def func2(X: np.ndarray | float):
    return 2* np.tan(X) - 5 * np.exp(X)
```

```python
def get_poly(
    deg: int,
    a: float | None = None,
    b: float | None = None,
    Xi: np.ndarray | None = None,
    Yi: np.ndarray | None = None,
):
    if Xi is None:
        Xi = np.linspace(a, b, deg + 1)

    A = np.ones((deg + 1, deg + 1))
    for i in range(1, deg + 1):
        A[:, i] = A[:, i - 1] * Xi.T

    return np.linalg.inv(A) @ Yi
```

```python
def poly_func(coeffs: np.ndarray, X: np.ndarray | float):
    if type(X) == float:
        res = 0
    else:
        res = np.zeros(X.shape)

    for i in range(len(coeffs)):
        res += coeffs[i]*(X**i)

    return res
```

```python
X = np.linspace(1, 6, 10_000)
Xi = np.array([1, 2, 3, 4, 5, 6])
Yi = np.array([14.5, 19.5, 30.5, 53.5, 94.5, 159.5])
poly = get_poly(5, Xi=Xi, Yi=Yi)

plt.plot(X, poly_func(poly, X), color = 'blue')

plt.scatter(Xi, Yi, c='blue')
poly_func(poly, 4.5)
```

```python
poly_1, Xi_1, Bi_1 = get_poly(2, -2, 2, func)
poly_2, Xi_2, Bi_2 = get_poly(3, -2, 2, func)
poly_3, Xi_3, Bi_3 = get_poly(4, -2, 2, func)
poly_4, Xi_4, Bi_4 = get_poly(5, -2, 2, func)
```

```python
figure, axis = plt.subplots(2, 2)

axis[0, 0].plot(X, poly_func(poly_1, X), color='blue')
axis[0, 0].scatter(Xi_1, Bi_1, c='blue')
axis[0, 0].plot(X, func(X), color = 'red')
axis[0, 0].set_title("degree = 2")

axis[0, 1].plot(X, poly_func(poly_2, X), color='blue')
axis[0, 1].scatter(Xi_2, Bi_2, c='blue')
axis[0, 1].plot(X, func(X), color = 'red')
axis[0, 1].set_title("degree = 3")

axis[1, 0].plot(X, poly_func(poly_3, X), color='blue')
axis[1, 0].scatter(Xi_3, Bi_3, c='blue')
axis[1, 0].plot(X, func(X), color = 'red')
axis[1, 0].set_title("degree = 4")

axis[1, 1].plot(X, poly_func(poly_4, X), color='blue')
axis[1, 1].scatter(Xi_4, Bi_4, c='blue')
axis[1, 1].plot(X, func(X), color = 'red')
axis[1, 1].set_title("degree = 5")
```

```python
X = np.linspace(-1.5, 1.5, 10_000)

poly_1, Xi_1, Bi_1 = get_poly(2, -1.5, 1.5, func2)
poly_2, Xi_2, Bi_2 = get_poly(3, -1.5, 1.5, func2)
poly_3, Xi_3, Bi_3 = get_poly(4, -1.5, 1.5, func2)
poly_4, Xi_4, Bi_4 = get_poly(5, -1.5, 1.5, func2)

figure, axis = plt.subplots(2, 2)

axis[0, 0].plot(X, poly_func(poly_1, X), color='blue')
axis[0, 0].scatter(Xi_1, Bi_1, c='blue')
axis[0, 0].plot(X, func2(X), color = 'red')
axis[0, 0].set_title("degree = 2")

axis[0, 1].plot(X, poly_func(poly_2, X), color='blue')
axis[0, 1].scatter(Xi_2, Bi_2, c='blue')
axis[0, 1].plot(X, func2(X), color = 'red')
axis[0, 1].set_title("degree = 3")

axis[1, 0].plot(X, poly_func(poly_3, X), color='blue')
axis[1, 0].scatter(Xi_3, Bi_3, c='blue')
axis[1, 0].plot(X, func2(X), color = 'red')
axis[1, 0].set_title("degree = 4")

axis[1, 1].plot(X, poly_func(poly_4, X), color='blue')
axis[1, 1].scatter(Xi_4, Bi_4, c='blue')
axis[1, 1].plot(X, func2(X), color = 'red')
axis[1, 1].set_title("degree = 5")
```
